# Support and Length

If we divide all items with row number, we can find Support value. Our threshold value is 0.2 for Support value.

```python
 #Calculate Product Frequency / Total Sales
first = pd.DataFrame(data.sum() / data.shape[0], columns=["Support"]).sort_values("Support", ascending=False)

# Format the values
first["Support"] = first["Support"].map("{:.2%}".format)

# Print the DataFrame using tabulate
print(tabulate(first, headers='keys', tablefmt='pretty', showindex=True, numalign='center'))
```

Elimination by Support Value 0.1

```python
# Calculate Product Frequency / Total Sales
first = pd.DataFrame(data.sum() / data.shape[0], columns=["Support"]).sort_values("Support", ascending=False)

# Filter based on support value >= 0.1
filtered_first = first[first.Support >= 0.1]

# Format the values
filtered_first["Support"] = filtered_first["Support"].map("{:.2%}".format)

# Print the filtered DataFrame using tabulate
print(tabulate(filtered_first, headers='keys', tablefmt='pretty', showindex=True, numalign='center'))
```

Second Iteration: Find support values for pair product combinations.

Elimination by Support Value 2

```python
second = list(itertools.combinations(first.index, 2))
second = [list(i) for i in second]
# Sample of combinations
second[:10]
```

```python
# Finding support values
value = []
for i in range(0, len(second)):
    temp = data.T.loc[second[i]].sum()
    temp = len(temp[temp == data.T.loc[second[i]].shape[0]]) / data.shape[0]
    value.append(temp)
# Create a data frame
secondIteration = pd.DataFrame(value, columns = ["Support"])
secondIteration["index"] = [tuple(i) for i in second]
secondIteration['length'] = secondIteration['index'].apply(lambda x:len(x))
secondIteration = secondIteration.set_index("index").sort_values("Support", ascending = False)
# Elimination by Support Value
secondIteration = secondIteration[secondIteration.Support > 0.1]
secondIteration
```

# Association Rules

# Association rules with 10% Support and 30% confidence

```python
# With 10% Support
frequent_itemsets=apriori(data,min_support=0.1,use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

# Insights:
Support value gives us these information:
Head 5

42 percent of 100 purchases are ChildBks
24 percent of 100 purchases are YouthBks
43 percent of 100 purchases are CookBks
28 percent of 100 purchases are DoltYBks
21 percent of 100 purchases are RefBks
Tail 5

10 percent of 100 purchases are ChildBks, GeogBks, DoItYBks
10 percent of 100 purchases are ChildBks, GeogBks, ArtBks
10 percent of 100 purchases are ArtBks, CookBks, DoItYBks
10 percent of 100 purchases are GeogBks, CookBks, DoItYBks
10 percent of 100 purchases are GeogBks, ArtBks, CookBks

```python
# with 30% Confidence
rules = association_rules(frequent_itemsets,metric='confidence', min_threshold=0.3)
rules
```

```python
rules.sort_values('lift',ascending=False)
```

```python
# Lift Ratio > 1 is a good influential rule in selecting the associated transactions
lift1=rules[rules.lift>1]
lift1
```

 Visualization of obtained rule

```python
# visualization of obtained rule
plt.figure(figsize=(16,9))
plt.scatter(rules['support'],rules['confidence'])
plt.xlabel('support')
plt.ylabel('confidence')
plt.show()
```

```python
matrix = lift1.pivot('antecedents','consequents','lift')
matrix
```

```python
plt.figure(figsize=(20,6),dpi=250)
sns.heatmap(matrix,annot=True)
plt.title('HeatMap - ForLiftMatrix')
plt.yticks(rotation=0)
plt.xticks(rotation=90)
```

```python
plt.figure(figsize=(16,9))
fit = np.polyfit(rules['lift'], rules['confidence'], 1)
fit_fn = np.poly1d(fit)
plt.plot(rules['lift'], rules['confidence'], 'yo', rules['lift'],
fit_fn(rules['lift']))
plt.xlabel('lift')
plt.ylabel('Confidence')
plt.title('lift vs Confidence')
```
