## Question 1

F&B manager wants to determine whether there is any significant difference in the diameter of the cutlet between two units. A randomly selected sample of cutlets was collected from both units and measured? Analyze the data and draw inferences at 5% significance level. Please state the assumptions and tests that you carried out to check validity of the assumptions

```python
import pandas as pd
import numpy as np
from scipy import stats
from scipy.stats import norm
```

```python
df1 = pd.read_csv("/content/Cutlets.csv")
```

```python
df1.head()
```

```python
da = df1["Unit A"]
db = df1["Unit B"]
```

```python
da.var()
```

```python
db.var()
```

```python
t_stats , P_value = stats.ttest_ind(da, db, equal_var=False)
```

```python
print(t_stats , P_value)
```

```python
aplha = 0.05
```

```python
if P_value < aplha :
  print("There is a significant difference between the two processes.")
else:
  print("There is no significant difference between the two process.")
```

## Question 2

A hospital wants to determine whether there is any difference in the average Turn Around Time (TAT) of reports of the laboratories on their preferred list. They collected a random sample and recorded TAT for reports of 4 laboratories. TAT is defined as sample collected to report dispatch.
   
  Analyze the data and determine whether there is any difference in average TAT among the different laboratories at 5% significance level.

```python
df2 = pd.read_csv("/content/LabTAT.csv")
```

```python
df2
```

```python
df2.describe()
```

```python
lab1,lab2,lab3,lab4 = df2["Laboratory 1"],df2["Laboratory 2"],df2["Laboratory 3"],df2["Laboratory 4"]
```

```python
aplha = 0.05
```

```python
F_sample , P_value1 = stats.f_oneway(lab1,lab2,lab3,lab4)
print(F_sample , P_value1 )
```

```python
if P_value1 < aplha:
  print("There is no difference in average TAT among the different laboratories")
else:
  print("There is a difference in average TAT among the different laboratories")
```

## Question 3

Sales of products in four different regions is tabulated for males and females. Find if male-female buyer rations are similar across regions.

```python
df03= pd.read_csv("/content/BuyerRatio.csv")
```

```python
df03.head()
```

```python
data = np.array([[50, 142, 131, 70], [435, 1523, 1356, 750]])
```

```python
chi2, p2, dof, expected = stats.chi2_contingency(data)
```

```python
aplha = 0.05
```

```python
print(f"Chi-Square Statistic: {chi2}")
print(f"P_value2: {p2}")
```

## Question 4

TeleCall uses 4 centers around the globe to process customer order forms. They audit a certain %  of the customer order forms. Any error in order form renders it defective and has to be reworked before processing.  The manager wants to check whether the defective %  varies by centre. Please analyze the data at 5% significance level and help the manager draw appropriate inferences


```python
import numpy as np
import pandas as pd
from scipy.stats import chi2_contingency
```

```python
df4 = pd.read_csv("/content/Costomer+OrderForm.csv")
```

```python
df4.head()
```

```python
contingency_table = pd.crosstab(df4['Phillippines'], [df4['Indonesia'], df4['Malta'], df4['India']])
```

```python
chi2, p4, dof, expected = chi2_contingency(contingency_table)
```

```python
print("Chi-Squared Statistic:", chi2)
print("P-value:", p4)
print("Degrees of Freedom:", dof)
print("Expected Frequencies Table:")
print(expected)
```

```python
alpha = 0.05
```

```python
if p4 > alpha:
    print("There is no significant association between call center and order form quality.")
else:
    print("There is a significant association between call center and order form quality.")
```
